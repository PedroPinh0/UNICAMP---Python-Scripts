```python
import numpy as np
import matplotlib.pyplot as plt

def E_Field(d,q1,q2):
    # d = Distância entre cargas
    # N = Escala do grid [N x N]
    # L = Tamanho do plano [L x L]
    
    N = 100
    L = 1   #[m]
    
    x = y = np.linspace(-L/2, L/2, N)
    X,Y   = np.meshgrid(x,y)
    
    R1  = ((X - d/2)**2 + Y**2)**0.5 
    R2  = ((X + d/2)**2 + Y**2)**0.5
    
    pot    = (q1/R1 + q2/R2)
    Ey, Ex = np.gradient(pot)
    
    c = np.log(np.hypot(Ex, Ey))
    plt.streamplot(x, y, Ex, Ey, color = c, linewidth=0.5, cmap=plt.cm.inferno, density=2, arrowstyle='->', arrowsize=1.5)
    plt.show()

d = 10e-2   #[m]
E_Field(d,1,1)
E_Field(d,1,-1)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from gaussxw import gaussxwab

def charge_dist(x,y):
    q0 = 100    #[Cm^-2]
    L  = 10e-2  #[m]
    return q0*np.sin(2*np.pi*x/L)*np.sin(2*np.pi*y/L)

def Potencial(x,y):
    # l = Tamanho do plano [l x l]
    # k = Constante de Coulomb

    k = 9e9   #[N m^2 C^-2]
    l = 1     #[m]
    
    Np = 20
    xp, wp = gaussxwab(Np, -l/2, l/2)
    
    I = 0.0
    for i in range(Np):
        for j in range(Np):
            r  = ((x - xp[i])**2 + (y - xp[j])**2)**0.5
            I += wp[i]*wp[j]*charge_dist(xp[i],xp[j])/r
    
    return I*k
    
a = 1e-2  #[m]
l = 1     #[m]
N = int(l/a)

x = y = np.linspace(-l/2,l/2, N) 
X,Y   = np.meshgrid(x,y)

Pot    = Potencial(X,Y)
Ey, Ex = np.gradient(Pot)
norm   = (Ex**2 + Ey**2)*0.5


plt.contourf(x,y,Pot, cmap = 'hsv')
plt.quiver(X,Y, Ex/norm, Ey/norm)
plt.show()
```
