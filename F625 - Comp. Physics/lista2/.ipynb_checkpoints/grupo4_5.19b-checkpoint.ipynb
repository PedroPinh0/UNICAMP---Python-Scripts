```python
import matplotlib.pyplot as plt
import numpy as np
from gaussxw import gaussxwab

# N: Número de fendas
# lambd: Comprimento de onda
# f: Distância focal
# L: Tamanho da tela
# a: Espaçamento entre fendas
# w: Tamanho total da grade de difração

N     = 10
lambd = 500e-9      #[m]
f     = 1           #[m]
L     = 10e-2       #[m]
a     = 20e-6       #[m]
w     = N*a

def q(u,a):
    return (np.sin(a*u))**2

def arg_fourier(x,u,lambd,f,a):
    i = complex(0,1)
    return abs(np.sqrt(q(u,a)))*np.exp(i*2*np.pi*x*u/(lambd*f))

def Interf_patt(x,lambd,f,a,w):
    n   = 100
    u,p = gaussxwab(n, -w/2, w/2)
    s     = 0.0
    for k in range(n):
        s += p[k]*arg_fourier(x,u[k],lambd,f,a)
        
    return (abs(s))**2

x = np.linspace(-L/2, L/2, 500)
I = Interf_patt(x,lambd,f,a,w)

plt.plot(x,I)
plt.show()
```

```python
N = 500
d = 0.0002

grid = np.empty([N,int(N/5)], np.float)

for i in range(N):
    x = d*(i - N/2)
    for j in range(int(N/5)):
        grid[i,j] = Interf_patt(x,lambd,f,a,w)
        
plt.imshow(grid,vmax=0.01,cmap='hot')
plt.gray()
plt.show()
```
