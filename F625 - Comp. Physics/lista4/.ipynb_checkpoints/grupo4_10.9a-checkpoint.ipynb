```python
import numpy as np
import matplotlib.pyplot as plt
from numpy.random import *

'''Simulao em estilo Metropolis do Modelo de
Ising para a magnetizao de um cristal 2D. As
constantes so tomadas em unidades naturais, onde
T = 1, k_b = 1 e J = 1.'''

# Constantes
rounds = int(1e6)   # Nmero de rodadas
N      = 20         # Dimenso da matriz
J      = 1          # Constante de Interao
T      = 1          # Temperatura
k_b    = 1          # Constante de Boltzmann
b      = 1/(k_b*T)

# Matriz onde ser armazenado os spins
S = (2*np.random.randint(0, 2, size = (N**2)) - 1).reshape(N,N)

def Energy(S):
    # Dimenso da matriz de spins
    N = np.shape(S)[0]
    
    # Clculo da interao entre vizinhos
    E1 = S[0:N-1,:]*S[1:N,:]   # Interao entre vizinhos verticais
    E2 = S[:,0:N-1]*S[:,1:N]   # Interao entre vizinhos horizontais

    # A energia total  a soma de todas as interaes de spins vizinhos
    E = np.sum(E1) + np.sum(E2)
    return -J*E

def Spin_Change(S):
    S_copy = S.copy()

    # Escolha aleatria de um spin
    i = randint(N)
    j = randint(N)
    
    # Clculo da energia antes e depois da mudana
    E_old = Energy(S_copy)
    S_copy[i,j] *= -1
    E_new = Energy(S_copy)
    
    # Verificao para ver se o movimento  ou no aceito
    dE = E_new - E_old
    if dE > 0:
        if random() < np.exp(-b*dE):
            return S_copy
        else:
            S_copy[i,j] *= -1
            return S_copy
    else:
        return S_copy
        
# Array de armazenamento da evoluo da magnetizao
M_plot = []
M = np.sum(S)

for k in range(rounds):
    S = Spin_Change(S)
    
    M += np.sum(S)
    
    M_plot.append(M/(k + 1))

plt.figure()
plt.plot(M_plot,'.k', markersize = 1)
plt.xlabel('Rodadas', size = 14)
plt.ylabel('Magnetizao [u.a]', size = 14)
plt.grid(alpha = 0.5)
plt.show()
```
